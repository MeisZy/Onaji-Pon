```python
import pandas as pd
```

```python
df = pd.read_csv('manga.csv')
df.head()
```

```python
def drop_columns(df):
    # Drop specified columns from the DataFrame
    columns_to_drop = ['anime_id', 'types', 'episodes', 'rating']
    df = df.drop(columns=columns_to_drop, errors='ignore')
    return df
```

```python
df = drop_columns(df)
df.head()
```
